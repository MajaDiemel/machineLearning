```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
value = np.loadtxt("vl5_data.csv")
tU = np.array(value[:,1])
tC = np.array(value[:,0])
plt.plot(tU, tC, "x")
```

```python
def lossFctn(tP, tC):
    squareDifference = (tP - tC)**2
    loss = squareDifference.mean()
    return loss

def dLossFctn(tP, tC):
    return 2*(tP - tC) / (np.size(tP))

def dModelDb(tU, w, b):
    return 1.0

def dModelDw(tU, w, b):
    return tU

def model(tU, b, w):
    tP = w*tU + b
    return tP

def gradFctn(tU, tP, tC, w,b):
    dLossDtp = dLossFctn(tP, tC)
    dLossDb = dLossDtp * dModelDb (tU, w, b)
    dLossDw = dLossDtp * dModelDw (tU, w, b)
    grad = np.stack([dLossDw.sum(), dLossDb.sum()])
    return grad

def training(nEpos, lRate, params, tU, tC):
    params = [params[0][0], params[1][0]]
    for i in range (0, nEpos):
        tP = model(tU, params[0], params[1])
        print(f'lossFunction = {lossFctn(tP, tC)}')
        grad = gradFctn(tU, tC, tP, params[0], params[1])
        params = params - lRate * grad
    return params

```

```python
w = np.ones(1)
b = np.zeros(1)
nEpos = 500
lRate = 1e-2
params = np.array([w, b])
tUn = tU * 0.1

param = training(nEpos, lRate, params, tU, tC)
print(param)
```
