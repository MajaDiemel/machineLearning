```python
import torch
import torch.nn as nn
from PIL import Image
from torchvision import transforms
import os
import matplotlib.pyplot as plt
```

```python
class ResNetBlock(nn.Module):
    def __init__(self, dim):
        super(ResNetBlock, self).__init__()
        self.conv_block = self.build_conv_block(dim)

    def build_conv_block(self, dim):
        conv_block = []

        conv_block += [nn.ReflectionPad2d(1)]

        conv_block += [nn.Conv2d(dim, dim, kernel_size=3, padding=0, bias=True),
                       nn.InstanceNorm2d(dim),
                       nn.ReLU(True)]

        conv_block += [nn.ReflectionPad2d(1)]

        conv_block += [nn.Conv2d(dim, dim, kernel_size=3, padding=0, bias=True),
                       nn.InstanceNorm2d(dim)]

        return nn.Sequential(*conv_block)

    def forward(self, x):
        out = x + self.conv_block(x) # <2>
        return out


class ResNetGenerator(nn.Module):
    def __init__(self, input_nc=3, output_nc=3, ngf=64, n_blocks=9): # <3> 

        assert(n_blocks >= 0)
        super(ResNetGenerator, self).__init__()

        self.input_nc = input_nc
        self.output_nc = output_nc
        self.ngf = ngf

        model = [nn.ReflectionPad2d(3),
                 nn.Conv2d(input_nc, ngf, kernel_size=7, padding=0, bias=True),
                 nn.InstanceNorm2d(ngf),
                 nn.ReLU(True)]

        n_downsampling = 2
        for i in range(n_downsampling):
            mult = 2**i
            model += [nn.Conv2d(ngf * mult, ngf * mult * 2, kernel_size=3,
                                stride=2, padding=1, bias=True),
                      nn.InstanceNorm2d(ngf * mult * 2),
                      nn.ReLU(True)]

        mult = 2**n_downsampling
        for i in range(n_blocks):
            model += [ResNetBlock(ngf * mult)]

        for i in range(n_downsampling):
            mult = 2**(n_downsampling - i)
            model += [nn.ConvTranspose2d(ngf * mult, int(ngf * mult / 2),
                                         kernel_size=3, stride=2,
                                         padding=1, output_padding=1,
                                         bias=True),
                      nn.InstanceNorm2d(int(ngf * mult / 2)),
                      nn.ReLU(True)]

        model += [nn.ReflectionPad2d(3)]
        model += [nn.Conv2d(ngf, output_nc, kernel_size=7, padding=0)]
        model += [nn.Tanh()]

        self.model = nn.Sequential(*model)

    def forward(self, input): # <3>
        return self.model(input)
    
    
netG = ResNetGenerator()
model_data = torch.load('horse2zebra_0.4.0.pth')
netG.load_state_dict(model_data)
netG.eval()
```

```python
horseList = []
folder = "horse/"
for horseImg in range(10):
    filepath = os.path.join(folder, "pferd" + str(horseImg) + ".jfif")
    img = Image.open(filepath)
    horseList.append(img)
```

```python
preprocess = transforms.Compose([transforms.Resize(256), transforms.ToTensor()])
batchList = []
for horseImg in horseList:
    img_t = preprocess(horseImg)
    batch_t = torch.unsqueeze(img_t, 0)
    batchList.append(batch_t)
```

```python
outList = []
for batch_t in batchList:
    batch_out = netG(batch_t)
    out_t = (batch_out.data.squeeze() + 1.0) / 2.0
    out = transforms.ToPILImage()(out_t)
    outList.append(out)
```

```python
# plot
fig, axs = plt.subplots(2, 10, figsize=(50, 10))
fig.suptitle("Horbras", y=0.9)

for i in range(10):
    axs[0, i].imshow(horseList[i])
    axs[0, i].axis('off')
    axs[1, i].imshow(outList[i])
    axs[1, i].axis('off')
plt.show()
```
